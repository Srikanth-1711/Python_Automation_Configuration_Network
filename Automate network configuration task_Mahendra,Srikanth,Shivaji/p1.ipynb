```python
'''

Project Title : Automate network configuration task

Modules Used  : json, os, platform, subprocess

Authors       : Mahendra, Srikanth, Shivaji


'''

import json
import os
import platform
import subprocess

def netInfo():
    '''
    Function to read network configuration, modify values, and store it in a new file.
    Also collects system network information and stores it in a JSON file.
    
    '''
    dict_ifcfg = {}  # Initialize an empty dictionary to store network configuration

    try:
        with open('C:\\Users\\polalaks\\Downloads\\ifcfg.conf', 'r') as f:
            for i in f:
                j = i.strip().split('=') # Split each line into key and value
                if len(j) == 2:  # Ensure there's both a key and value
                    dict_ifcfg[j[0]] = j[1]
                else:
                    print(f"Skipping malformed line: {i.strip()}")              
        print(dict_ifcfg)
    except FileNotFoundError:
        print("File not found. Please check the file path.")
    except Exception as e:
        print(f"An error occurred: {e}")

    # Modify the configuration dictionary with predefined values
    dict_ifcfg['onboot']='yes' 
    dict_ifcfg['bootproto']='static'
    dict_ifcfg['defroute']='no'
    print('After Modification')
    print(dict_ifcfg)

    # Further modify the dictionary to add IP address and subnet mask
    dict_ifcfg['IP Address']='10.1.1.1'
    dict_ifcfg['subnet']='255.255.255.0'
    
    print(dict_ifcfg) 

    try:
        new_f=open('net_ifcfg','w')
        for i,j in dict_ifcfg.items():
            new_f.write(f'{i} = {j}'+'\n')
        new_f.close()

    except FileNotFoundError:
        print("File not found. Please check the file path.")
    except Exception as e:
        print(f"An error occurred: {e}")
    
    with open('net_ifcfg.json', 'w') as new_jf:
        json.dump(dict_ifcfg, new_jf, indent=4)
    
    os_name = platform.system()
    print(os_name)

    netInfo={} 
    netInfo['Windows']=[]  # Prepare a list for macOS commands


    def exe_comm(comm):
        '''Function to execute a system command and return its output'''
        
        return subprocess.check_output(comm, shell=True, text=True) # Execute the command and return the output
    
    
    netInfo['Windows'].append(exe_comm('powershell Get-Netroute'))
    netInfo['Windows'].append(exe_comm('powershell Get-NetIPAddress'))

    with open('net_ifcfg.json', 'r') as json_file:
        existing_data = json.load(json_file)
    
    
   
    existing_data.update(netInfo)
    print(existing_data)


    with open('net_ifcfg.json', 'w') as new_jf:
        json.dump(existing_data,new_jf, indent=4)



if __name__=='__main__':
    netInfo()
```
